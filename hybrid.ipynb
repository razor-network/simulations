```python
import matplotlib.pyplot as plt
import math
import random
from numpy.random import default_rng
import numpy as np
```

```python

stakers=[]
numStakers=10
for i in range(1,numStakers+1):
    stakers.append(10000*i*i)
    
print(stakers)
stakersBefore = stakers.copy()
# print(sum(stakers))
# plt.figure(0)

# plt.plot(stakers)
epochs=1000
dynastyEpochs=1
exposure=1000
minStake=1000
roll=0.05
tolerance = 0
blockReward = 0
#concentrationPenalty = 10000
age =  np.array([1 for i in range(1,numStakers+1)], dtype=np.float64)

print(age)

stakes=[stakers]
    
print(stakes)

def vote():
    return default_rng().normal(loc=100, scale=10)
    
def penalty(stake):
    return stake/exposure

def chooseblockProposer(stakers):
    winner = math.floor(random.randrange(math.floor(sum(stakers))))
    #print(winner)
    sortedStakers=sorted(stakers)    
    agg=0
    for i in range(len(stakers)):
        agg+=sortedStakers[len(stakers)-i-1]
        if agg>winner:
            return len(stakers)-i-1
        
def getMaturity(age):
    return min(math.log(age+10000) - math.log(10000),0.05)
print(getMaturity(10000))
```

```python
winners=[0]*numStakers
influences=stakers.copy()
influencesOverEpochs=[]
agesOverEpochs=[]

def main():
    rewardPool=0
    rewards=[0]*numStakers

    for i in range(epochs):

        votes=[]

        ## VOTE ##
        for j in range(len(stakers)):
            if stakers[j]<minStake:
                votes.append(0)
                continue
            votes.append(vote())

        #print('votes',votes)
        uVotes=set(votes)
        #print('uVotes',uVotes)
        ## WeightedVotes ##
        wVotes={}
        weights=[]

        for j in uVotes:
            weight = 0
            if j==0:
                continue
            for k in range(len(votes)):

                if j==votes[k]:
                    weight+=influences[k]
                wVotes[j]=weight
        #print('wVotes',wVotes)
        wVotesSorted=sorted(wVotes) #list
        #print('wVotesSorted',wVotesSorted)
                  
        totalInfluence=sum(influences)
        medianWeight = totalInfluence/2
        lowerCutoffWeight = totalInfluence/4
        higherCutoffWeight = (totalInfluence*3)/4
        lowerCutoff = 0
        higherCutoff = 0
        median = 0
        agg=0
        #print('cutoffWeights',lowerCutoffWeight,medianWeight,higherCutoffWeight)
        for k in ((wVotesSorted)):
            agg+=wVotes[k]
            #print('agg',agg)
            if lowerCutoff == 0 and agg >= lowerCutoffWeight:
                lowerCutoff = k
            if median == 0 and agg >= medianWeight:
                median = k    
            if higherCutoff == 0 and agg >= higherCutoffWeight:
                higherCutoff = k
        #print("cutoffs", lowerCutoff, median, higherCutoff)

        ## PENALTY ##
        stakeGettingReward=0
        for j in range(len(stakers)):
            if stakers[j]<minStake:
                continue
            if (votes[j]<lowerCutoff or votes[j]>higherCutoff):
                penalty = ((median-votes[j])**2)/median**2
                #print('penalizing', votes[j], lowerCutoff,higherCutoff, median, penalty)

                #print("penalty", penalty)
                # penalty cant be more than 100%
                if(penalty>1): penalty = 1
                    
#                 print(influences[j])
#                 rewardPool+=penalty
            else:
                penalty=0

            #print('age before', age[j])
            age[j] = age[j] + 1.0 - age[j]*penalty
            #print('age after', age[j])
            #print('influence before', influences[j])
            influences[j] = getMaturity(age[j])*stakers[j]
            #print('influence after', influences[j])
        #print('ages this epoch', age)



#                 rewards[j]+=1

        ## BLOCKREWRD ##
        winner = chooseblockProposer(stakers)
        #print(winner)
        winners[winner]+=1
        #print('winners',winners)
        #print(stakers)
        stakers[winner]+=blockReward
        #print(stakers)


        stakes.append(stakers)
        influencesOverEpochs.append(influences.copy())
        #print('influences',influences)
        agesOverEpochs.append(age.copy())
        #print('agesOverEpochs',agesOverEpochs)


    
    
main()
print('don)')

#print('agesOverEpochs',(agesOverEpochs))
plt.plot(agesOverEpochs)
plt.suptitle('ageOverEpochs')
plt.figure(2)

#print('influencesOverEpochs',(influencesOverEpochs))
plt.plot(influencesOverEpochs)
plt.suptitle('influencesOverEpochs')
plt.figure(3)

print((stakers))
print((stakersBefore))

for i in range(len(stakers)):
    print(stakers[i]/stakersBefore[i])
```
