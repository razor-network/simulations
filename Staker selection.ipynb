```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random
numStakers=10000
difficulty=5
iterations = 1000
results = []
stakers = [random.random()*i**2 for i in range(numStakers)]
```

```python
stakers
plt.plot(stakers)
```

```python
maxStake = max(stakers)
```

```python
maxStake
```

```python
for iteration in range(iterations):
    selected = 0
    lucky=[]
    for i in stakers:
        if(random.random()<i/(maxStake*difficulty)):
            selected+=1
            lucky.append(i)
    print(difficulty,selected)
    results.append([iteration,difficulty,selected])
    if selected<1000:
        difficulty*=0.95
    else:
        difficulty*=1.05
```
