```python
#according to whitepaper version 0.4
import matplotlib.pyplot as plt
import math
import random
from numpy.random import default_rng
import numpy as np
```

```python

stakers=[]
numStakers=10
for i in range(1,numStakers+1):
    stakers.append(10000*i*i)
    
print(stakers)
stakersBefore = stakers.copy()
# print(sum(stakers))
# plt.figure(0)

# plt.plot(stakers)
epochs=10000
dynastyEpochs=1
exposure=1000
minStake=1000
roll=0.05
tolerance = 0
blockReward = 0
#concentrationPenalty = 10000
age =  np.array([1 for i in range(1,numStakers+1)], dtype=np.float64)

print('age',age)

stakes=[stakers]
    
print(stakes)

def vote():
    return default_rng().normal(loc=100, scale=10)
    
def penalty(stake):
    return stake/exposure

def chooseblockProposer(stakers):
    winner = math.floor(random.randrange(math.floor(sum(stakers))))
    sortedStakers=sorted(stakers)    
    agg=0
    for i in range(len(stakers)):
        agg+=sortedStakers[len(stakers)-i-1]
        if agg>winner:
            return len(stakers)-i-1
        
def getMaturity(age):
    return min(math.log(age+10000) - math.log(10000),9999)
print(getMaturity(10000))
print(len(stakers))
```

```python
winners=[0]*numStakers
influences=[]
maturities=[]
for i in range(len(stakers)):
    inf = stakers[i]*getMaturity(age[i])
    influences.append(inf)

for i in range(len(stakers)):
    mat = getMaturity(age[i])
    #print('inf',inf)
    maturities.append(mat)

print('influences',influences)
influencesOverEpochs=[]
maturitiesOverEpochs=[]

agesOverEpochs=[]
resultsOverEpochs=[]

def main():
    rewardPool=0
    rewards=[0]*numStakers

    for i in range(epochs):

        votes=[]

        ## VOTE ##
        for j in range(len(stakers)):
            if stakers[j]<minStake:
                votes.append(0)
                continue
            votes.append(vote())

        uVotes=set(votes)
        ## WeightedVotes ##
        wVotes={}
        weights=[]

        for j in uVotes:
            weight = 0
            if j==0:
                continue
            for k in range(len(votes)):

                if j==votes[k]:
                    weight+=influences[k]
                wVotes[j]=weight
        wVotesSorted=sorted(wVotes) #list
                  
        totalInfluence=sum(influences)
        medianWeight = totalInfluence/2
        median = 0
        agg=0
        for k in ((wVotesSorted)):
            agg+=wVotes[k]

            if median == 0 and agg >= medianWeight:
                median = k    

        resultsOverEpochs.append(median)
        ## PENALTY ##
        stakeGettingReward=0
        penalty=0
        for j in range(len(stakers)):
            if stakers[j]<minStake:
                continue
            if votes[j]==0:
                penalty=1
            else:
                penalty = ((median-votes[j])**2)/median**2
                # penalty cant be more than 100%
                if(penalty>1): penalty = 1
                    

            age[j] = min(100, age[j] + 1.0 - age[j]*penalty)
            maturity=getMaturity(age[j])
            maturities[j]=maturity
            influences[j] = maturity*stakers[j]

        ## BLOCKREWRD ##
        winner = chooseblockProposer(stakers)
        winners[winner]+=1
        stakers[winner]+=blockReward

        stakes.append(stakers)
        influencesOverEpochs.append(influences.copy())
        agesOverEpochs.append(age.copy())
        maturitiesOverEpochs.append(maturities.copy())


    
    
main()
print('done')
plt.figure(figsize=(20, 9), dpi=80)

plt.plot(agesOverEpochs)
plt.suptitle('ageOverEpochs')
plt.figure(2)

plt.figure(figsize=(20, 9), dpi=80)

plt.plot(influencesOverEpochs)
plt.suptitle('influencesOverEpochs')
plt.figure(3)

plt.figure(figsize=(20, 9), dpi=80)


plt.plot(maturitiesOverEpochs)
plt.suptitle('maturitiesOverEpochs')
plt.figure(4)


plt.plot(resultsOverEpochs)
plt.suptitle('resultsOverEpochs')
plt.figure(5)

print((stakers))
print((stakersBefore))

for i in range(len(stakers)):
    print(stakers[i]/stakersBefore[i])
```
